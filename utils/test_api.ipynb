```python
import requests

BASE_URL = "http://localhost:8000"


def scrape(channel_url: str, person: str, date_from: str, date_to: str) -> None:
    print(f"\n{'─' * 60}")
    print(f"Channel: {channel_url}")
    print(f"Person : {person}")
    print(f"Period : {date_from} → {date_to}")
    print("─" * 60)

    resp = requests.post(
        f"{BASE_URL}/scrape",
        json={
            "channel_url": channel_url,
            "person":      person,
            "date_from":   date_from,
            "date_to":     date_to,
        },
        timeout=300,
    )

    if resp.status_code != 200:
        print(f"[ERROR] {resp.status_code}: {resp.text}")
        return

    data = resp.json()

    if data["status"] == "exists":
        print(f"[EXISTS] Data already in DB — {data['total_in_range']} posts")
    else:
        print(f"[ADDED]  Added {data['newly_added']} new posts "
              f"(total in range: {data['total_in_range']})")

    print(f"\nFirst post : {data['posts'][0]['posted_at']}  {data['posts'][0]['text'][:80]}")
    print(f"Last post  : {data['posts'][-1]['posted_at']}  {data['posts'][-1]['text'][:80]}")
    return data

data = scrape(
        channel_url="t.me/vereshchuk_iryna",
        person="Iryna Vereshchuk",
        date_from="2023-12-01T00:00:00Z",
        date_to="2024-02-01T00:00:00Z",
    )
data
```
